```python
group1 = df.groupby('Team 1').size().rename('Team A')
group2 = df.groupby('Team 2').size().rename('Team B')
total_df = pd.concat([group1,group2],axis =1 ).fillna(0)
total_df['Total'] = total_df['Team A']+ total_df['Team B']
total_df.sort_values(by=['Total'],ascending= False).head(4)
```
