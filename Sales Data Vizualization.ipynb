# We'll be analysing  Sales Report

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_excel('sales.xlsx')
df.head()
```

```python
sns.relplot(x='Sales',y='Profit',data=df, hue= 'Ship Mode', style = 'Order Priority')
```

## We developed to scratter plot between sales & profit
and we have noticed that we even have negative profit sales as well.
Orange color is most dominating here & this gives a message to Logistics team
that most of the orders delivered through the Delivery Truck 

```python
sns.relplot(x='Sales',y='Profit',data=df, col= 'Ship Mode', hue= 'Product Category',col_wrap=3)
```

```python
sns.relplot(x='Sales',y='Profit',data=df, col= 'Customer Segment',row='Product Category',hue = 'Ship Mode')
```

# Here we looked into each product category for each customer segment
Findings through this chart:
1. Most of the Furniture category got transfered through Delivery Truck
2. Most of the office supplies got transfered through Express Air
3. Technology got transfered almost equality with Delivery Truck & Regular air 

```python
sns.catplot(x='Ship Mode', y='Sales',data=df.query('Profit>1000'),jitter=True)
```

# Scatter Plot where profit is greater than 1000 with shipping modes

```python
sns.relplot(x='Order Date', y='Sales', data= df, kind ='line')
```

# Now we will be analysing data with the time period 
As we get messy timeline we need to do some data cleaning

```python
df['year']= df['Order Date'].dt.year
df['year'].unique()
```

```python
sns.relplot(x='year', y='Sales', data=df, kind='line')
```

```python
sns.relplot(x='year', y='Sales', data=df, kind='line',ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=df, kind='line',hue='Ship Mode',col='Product Category',ci=None)
```

```python
sns.relplot(x='year', y='Sales', data=df, kind='line',hue='Ship Mode',col='Region',ci=None, col_wrap=4)
```
